# Exercícios Aula 2

```python
print("Somando números")
try:
    num_1 = float(input('Digite o primeiro valor: '))
    num_2 = float(input('Digite o segundo valor: '))
    soma = num_1 + num_2

except (ValueError):
        print('''Por favor digite apenas números. 
        \nFim do programa. 
        \nExecute novamente para continuar.''')
else:
    print(f"A soma: {num_1} + {num_2} = {soma}")
    print('Fim do programa.')
```

```python
print('Calculadora de segundos')

try:
    dia = int(input('Digite o número  de dias: '))
    horas = int(input('Digite o número  de horas: '))
    min = int(input('Digite o número  de minutos: '))
    seg = int(input('Digite o número  de segundos: '))
    seg_result = ((dia * 86400) + (horas * 3600) + (min * 60) + seg)

except (ValueError):
        print('''Por favor digite apenas número inteiros. 
        \nFim do programa. 
        \nExecute novamente para continuar.''')
else:
    print(f'{dia} Dias, {horas} horas, {min} minutos, {seg} segundos, resulta em: {seg_result} segundos.')
    print('Fim do programa.')
```

```python
print("CALCULADORA DE DESCONTOS!")
try:
    preco = float(input('Digite o preco: '))
    percentual_desconto = float(input('Digite a porcentagem do desconto: '))
    novo_preco = (preco * percentual_desconto) / 100
except (ValueError):
        print('''Por favor digite apenas números. 
        \nFim do programa. 
        \nExecute novamente para continuar.''')
else:
    print(f'O seu produto no valor de R${preco} \nCom o desconto de {percentual_desconto}% \nRecebe o novo preço de: R${novo_preco}')
```
